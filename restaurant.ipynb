```
sales_data = [
          ('A', '2021-01-01', '1'),
          ('A', '2021-01-01', '2'),
          ('A', '2021-01-07', '2'),
          ('A', '2021-01-10', '3'),
          ('A', '2021-01-11', '3'),
          ('A', '2021-01-11', '3'),
          ('B', '2021-01-01', '2'),
          ('B', '2021-01-02', '2'),
          ('B', '2021-01-04', '1'),
          ('B', '2021-01-11', '1'),
          ('B', '2021-01-16', '3'),
          ('B', '2021-02-01', '3'),
          ('C', '2021-01-01', '3'),
          ('C', '2021-01-01', '1'),
          ('C', '2021-01-07', '3')]
# cols
sales_cols= ["customer_id", "order_date", "product_id"]
sales_df = spark.createDataFrame(data = sales_data, schema = sales_cols)
# view the data
sales_df.show()
```

```
menu_data = [ ('1', 'palak_paneer', 100),
              ('2', 'chicken_tikka', 150),
              ('3', 'jeera_rice', 120),
              ('4', 'kheer', 110),
              ('5', 'vada_pav', 80),
              ('6', 'paneer_tikka', 180)]
# cols
columns=['product_id', 'product_name', 'price']
menu_df= spark.createDataFrame(data = menu_data, schema = columns)
menu_df.show()
```

```
# create member data
members_data = [ ('A', '2021-01-07'),
                 ('B', '2021-01-09')]
columns=['customer_id','joining_data']
members_df= spark.createDataFrame(data = members_data, schema= columns)
members_df.show()
```

```
total_spent_df = (sales_df.join(menu_df, 'product_id') 
                         .groupBy('customer_id').agg({'price':'sum'}) 
                         .withColumnRenamed('sum(price)', 'total_spent_amounts') 
                         .orderBy('customer_id'))
total_spent_df.show()
```

```
days_visited= sales_df.groupBy('customer_id').agg(countDistinct('order_date'))\
               .withColumnRenamed('count(order_date)', 'Times_visited')
               
days_visited.show()
```

```
from pyspark.sql.functions import *
res_visit_df = sales_df.groupBy('customer_id').agg(countDistinct('order_date')).withColumnRenamed('count(order_date)', 'visit_days').orderBy('customer_id')
res_visit_df.show()
```
