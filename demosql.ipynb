```
schema1 = StructType().add("date",StringType(),True).add("time",StringType(),True).add("size",DoubleType(),True).add("r_version",StringType(),True).add("r_arch",StringType(),True).add("r_os",StringType(),True).add("package",StringType(),True).add("version",StringType(),True).add("country",StringType(),True).add("ip_id",DoubleType(),True)
```

```
dataDF = spark.read.format("csv").option("header","TRUE").schema(schema1).load("/FileStore//tables/2015_12_12.csv")
```
