```
for i in arrayrdd.take(5):
    print(i)
```

```
key = arrayrdd.map(lambda a : (a[0],float(a[2])*int(a[3])))

total = key.reduceByKey(lambda a,b : a+b)
total1=total.sortBy(lambda a : -a[1])

for i in total1.take(5):
	print(i)
```

```
key=arrayrdd.map(lambda a : (a[0],int(a[3])))
total=key.reduceByKey(lambda a,b : a+b)
total1=total.sortBy(lambda a : -a[1])

for i in total1.take(1):
    print("The year with most number of passenger is",i[0],"with",i[1],"passenger")
```
