```

from pyspark.sql.types import StructType, StructField, IntegerType, StringType, LongType
from pyspark.sql.functions import *
```

```
files_loc= '/FileStore/tables/D*'
Filetype='csv'
delimeter= '\;'
```

```
retail_schema= StructType([
    StructField('Txn_dt' , StringType(), nullable=True),
    StructField('Custno' , StringType(), nullable=True),
    StructField('Age' , StringType(), nullable=True),
    StructField('Zipcode' , StringType(), nullable=True),
    StructField('Category' , StringType(), nullable=True),
    StructField('Product' , StringType(), nullable=True),
    StructField('Qty' , IntegerType(), nullable=True),
    StructField('Cost' , LongType(), nullable=True),
    StructField('Sales' , LongType(), nullable=True),
    
])
```

```
retail_data= spark.read.format(Filetype).option("sep",';').option("inferSchema",True).schema(retail_schema).load(files_loc)
```

```
display(retail_data)
```

```
retail_data.registerTempTable("retail_table")
```

```


# df1=spark.sql("select age, count(distinct(Custno)) as No_of_customer, sum(sales) as Total_Sales from retail_table where month(Txn_dt)=1 group by Age order by Total_Sales desc ")
from pyspark.sql.functions import * 
df1=retail_data.select('Age', 'Custno', 'sales')\
    .groupBy('Age')\
    .agg(countDistinct("Custno"), sum("sales").alias("Total_Sales"))\
    .orderBy('Total_Sales', ascending=False)
df1.show(10)
```

```
#df2=spark.sql("select Product ,count(distinct(Custno) )as Number_of_Customer, sum(sales) as Total_sales from retail_table where trim(Age)= 'A' group by Product order by Total_Sales desc limit 10 ")


df = retail_data.filter(trim(retail_data.Age) == "A") \
                 .groupBy("Product") \
                 .agg(countDistinct("Custno").alias("Number_of_Customer"), \
                      sum("sales").alias("Total_sales")) \
                 .orderBy("Total_sales", ascending=False) \
                 .limit(10)

df.show()
```

```
#area_wise_sales=spark.sql("select Zipcode, sum(sales) as Total_sales from retail_table group by Zipcode order by Total_sales desc limit 10")

df3= retail_data.select('Zipcode','sales')\
    .groupBy('Zipcode')\
    .agg(sum('sales').alias("Total_Sales"))\
    .orderBy('Total_Sales', ascending=False)\
    .limit(10)
df3.show()
```

```
df5=retail_data.select('Product','sales','cost')\
                .groupBy('Product')\
                .agg(sum(retail_data.Sales - retail_data.Cost).alias('profit_loss'))\
                .orderBy('profit_loss',ascending = False)\
                .limit(10).show()
                
```

```
df4=retail_data.select('Product','sales','cost')\
                .groupBy('Product')\
                .agg(sum(retail_data.Sales - retail_data.Cost).alias('profit_loss'))\
                .filter(col('profit_loss')< 0)\
                .orderBy('profit_loss',ascending = True)\
                .limit(10).show()
```
